#### Task2: Descriptive Statistics and Visualizations

## BOX PLOT

```python
pyplot.figure(figsize=(8, 6))
sns.boxplot(data=boston_df, y='MEDV', color='lightblue')
# Add labels and title
pyplot.title('Boxplot of Median Value of Owner-Occupied Homes', fontsize=16)
pyplot.xlabel('Median Value', fontsize=12)
pyplot.show()
```

## BAR PLOT

```python
pyplot.figure(figsize=(8, 6))
sns.barplot(data=boston_df, y='CHAS', color='lightblue', palette='pastel')
# Add labels and title
pyplot.title('BAR PLOT of Charles River', fontsize=16)
pyplot.xlabel('Charles River', fontsize=12)
pyplot.show()
```

## BOX PLOT for MEDV vs AGE Variable

```python
bins = [0, 35, 70, 100]  # Age ranges
labels = ['35 years and younger', '35-70 years', '70 years and older']
boston_df['AGE_Group'] = pd.cut(boston_df['AGE'], bins=bins, labels=labels, right=True)

pyplot.figure(figsize=(10, 6))
sns.boxplot(data=boston_df, x='AGE_Group', y='MEDV', palette='pastel')

pyplot.title('Boxplot of MEDV by Age Group', fontsize=16)
pyplot.xlabel('Age Group', fontsize=12)
pyplot.ylabel('Median Value (in $1000)', fontsize=12)
pyplot.show()
```

## SCATTER PLOT for MEDV vs AGE Variable

```python
pyplot.figure(figsize=(8, 6))
pyplot.scatter(boston_df['INDUS'], boston_df['NOX'], alpha=0.7, color='blue')

# Add labels and title
pyplot.title('Scatter Plot of NOX vs. INDUS', fontsize=16)
pyplot.xlabel('Proportion of Non-Retail Business Acres per Town (INDUS)', fontsize=12)
pyplot.ylabel('Nitric Oxide Concentrations (NOX)', fontsize=12)

# Show the plot
pyplot.grid(alpha=0.3)
pyplot.show()
```

## HISTOGRAM FOR THE PUPIL

```python
pyplot.figure(figsize=(8, 6))
pyplot.hist(boston_df['PTRATIO'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)

pyplot.title('Histogram of Pupil-to-Teacher Ratio (PTRATIO)', fontsize=16)
pyplot.xlabel('Pupil-to-Teacher Ratio', fontsize=12)
pyplot.ylabel('Frequency', fontsize=12)

pyplot.grid(alpha=0.3)
pyplot.show()
```
